### The simplest plot using pyplot

```python
import matplotlib.pyplot as plt

POINTS = 10
```

```python
x_numbers = [i for i in range(POINTS)]
y_numbers = [i * 2 for i in x_numbers]

plt.plot(x_numbers, y_numbers)
plt.show()
```

```python
points = [(i, i**2) for i in range(10)]

plt.plot(points)
plt.show()
```
