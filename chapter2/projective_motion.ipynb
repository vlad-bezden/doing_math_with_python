```python
from matplotlib import pyplot as plt
import math
```

```python
'''
Draw the trajectory of a body in porjective motion
'''
def draw_graph(x, y):
    plt.plot(x, y)
    plt.xlabel('meters')
    plt.ylabel('meters')
    plt.title('Projective motion of a ball')
```

```python
'''
Generate equally spaced floating point
numbers between two given values
'''
def frange(start, final, increment):
    while start < final:
        yield start
        start += increment
```

```python
def draw_trajectory(u, theta):
    theta = math.radians(theta)
    g = 9.8
    
    # time of flight
    t_flight = 2*u*math.sin(theta)/g
    # find time intervals
    intervals = frange(0, t_flight, 0.001)
    
    # list of x and y coordinates
    x = []
    y = []
    
    for t in intervals:
        x.append(u*math.cos(theta)*t)
        y.append(u*math.sin(theta)*t-0.5*g*t**2)
        
    draw_graph(x, y)
```

```python
try:
    u = float(input('Enter the initial velocity (m/s): '))
    theta = float(input('Enter the angle of projection (degrees): '))
except ValueError:
    print('You entered an invalid input')
else:
    draw_trajectory(u, theta)
    plt.show()
```

```python
# baseball pitcher for different group can throw
# ball for:
# 10U          -> 40-50mph (18-22ms)
# 11-12        -> 50-60mph (22-27ms)
# 13-14        -> 55-75mph (25-34ms)
# HS           -> 75-85mph (34-38ms)
# Collage/Pro  -> 80-95mph (36-42ms)
u_list = [('10U', 20), ('11-12', 25), ('13-14', 28), ('HS', 36), ('Collage/Pro', 42)]
theta = 45

for _, v in u_list:
    draw_trajectory(v, theta)
    
# add a legend and show the graph
plt.legend([l for l, _ in u_list])
plt.show()
```
